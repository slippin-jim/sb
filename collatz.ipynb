```python
import random
import pandas as pd
import plotly.express as px
```

```python
def collatz(n, result=None, print_=False):
    if result == None:
        result = []
    if n == 1:
        result.append(1)
        print(n) if print_ == True else ''
    elif n % 2 == 0:
        result.append(n)
        print(n) if print_ == True else ''
        collatz(int(n/2),result)
    else:
        result.append(n)
        print(n) if print_ == True else ''
        collatz(int((n*3))+1,result)
    return result
```

```python
collatz(35,print_=True)
```

```python
i = 0
v = []
c = []
y = []
while i < 2000:
    x = random.randint(1,1000)
    _ = collatz(x)
    v += _
    counter = 0
    c += [i for i,z in enumerate(_)]
    y += [x]*len(_)
    i+=1
```

```python
p = pd.DataFrame(data=zip(v,c,y),columns=['v','c','y'])
px.line(p,x='c',y='v',color='y',log_x=True)
```
