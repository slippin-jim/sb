```python
#!/Users/jim/anaconda3/envs/sb/bin/python

import pandas as pd
import time
import os
from datetime import datetime as dt
import logging
```

```python
pga_results = pd.read_csv('~/dev/sb/data/pga_stats.csv')
```

```python
pga_results['event_completed'] = pd.to_datetime(pga_results.event_completed)
pga_results.set_index('event_completed',inplace=True)
pga_results.head()
```

```python
pga_results['2Y_sg'] = pga_results.groupby('dg_id')['sg_total'].transform(lambda d: d.rolling('720D', min_periods=1).mean())
pga_results['2M_sg'] = pga_results.groupby('dg_id')['sg_total'].transform(lambda d: d.rolling('60D', min_periods=1).mean())
pga_results
```
